```python
# Import Dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
sonar_data = pd.read_csv("sonar_data.csv",header=None)
sonar_data.head()
```

```python
sonar_data.describe()
```

```python
sonar_data.shape
```

```python
sonar_data[60]
```

```python
sonar_data[60].value_counts()
```

```python
# Mean Values for Decision

sonar_data.groupby(60).mean()
```

```python
# Seperate Data and Labels

X = sonar_data.drop(columns=60,axis=1)
y = sonar_data[60]
```

```python
print(X)
```

```python
print(y)
```

```python
# Training and Test Data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)
```

```python
print(X_train)
```

```python
# Logistic Regression Model Training

model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
# Model Evaluation

#accuracy of training data
X_train_prediction = model.predict(X_train)
training_model_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print("Accuracy of the Training data :{}".format(training_model_accuracy))
```

```python
#accuracy of test data
X_test_prediction = model.predict(X_test)
test_model_accuracy = accuracy_score(X_test_prediction, y_test)

print("Accuracy of the Test data :{}".format(test_model_accuracy))
```
